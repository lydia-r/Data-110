```python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
iris=sns.load_dataset('iris') # loading dataset from seaborn
```

```python
iris.head()
```

```python
iris.info()
```

# 1. Data Visualization

### 1.1 Pair Plot

```python
sns.pairplot(iris) # Pairplot will create all scatter plots for the numerical columns
```

```python
# Pairplot using a hue for species
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset')
plt.show()
```

### 1.2 Bar Plot

```python
species_mean= iris.groupby('species')['sepal_length'].mean() # grouping by species and using mean to show average sepal length
plt.bar(species_mean.index, species_mean.values)
plt.ylabel('Sepal length')
plt.title('Average sepal length of each species')
```

### 1.3 Box Plot

```python
# Box Plot for petal width distribution by species
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Distribution of Petal Width by Species')
plt.show()
```

### 1.4 Grouped Bar Plot


```python
# Grouping the dataset by species and calculating mean for both sepal length and petal length
iris_grouped = iris.groupby('species').mean()[['sepal_length', 'petal_length']].reset_index()

iris_grouped.plot(x='species', kind='bar', figsize=(8,6))
plt.title('Average Sepal and Petal Lengths by Species')
plt.ylabel('Length (cm)')
plt.xticks(rotation=0)
plt.show()
```

### 1.5 Violin Plot

```python
sns.violinplot(x='species', y='sepal_length', data=iris) # ploting violin graph to show the distribution of each species
plt.title('Distribution of Sepal length by species')
plt.show()
```
