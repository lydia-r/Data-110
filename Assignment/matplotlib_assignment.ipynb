1. Setup and Style


```python
# import the libraries
import matplotlib.pyplot as plt
import numpy as np

print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
alternative_style ='Solarize_Light2'
```

2. Basic Line Plot

```python
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1

plt.plot(x, y)
plt.title('Basic Line Plot')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

3. Multiple lines

```python
x = np.linspace(1, 4, 100)
x_log = np.linspace(1, 4, 100)  # log function cannot handle x <= 0

f_x = x**3
g_x = np.exp(x)
h_x = np.log(x_log)

plt.plot(x, f_x, label='$f(x) = x^3$', color='yellow')
plt.plot(x, g_x, label='$g(x) = e^x$', color='green')
plt.plot(x_log, h_x, label='$h(x) = \log(x)$', color='red')

plt.legend(bbox_to_anchor=(1, 1))
plt.title('Multiple Functions Plot')
plt.xlabel('Function Values')
plt.show()
```

4. Styling Lines

```python
x = np.linspace(-2, 4, 100)

plt.plot(x, x**2, '-g', label='$f(x) = x^2$') # Solid green
plt.plot(x, x**3, '--c', label='$g(x) = x^3$') # dashed cyan
plt.plot(x, x**4, '-.k', label='$h(x) = x^4$') # dashdot black
plt.plot(x, x**5, ':r', label='$i(x) = x^5$') # dotted red

plt.legend(bbox_to_anchor=(1, 1))
plt.title('Different Styling Lines')
plt.xlabel('x')
plt.ylabel('Values')
plt.show()
```

5. Axes Limits and Scaling

```python
x = np.linspace(-10, 10, 200)
f_x = 1 / (x**2 + 1)

plt.plot(x, f_x)
plt.xlim(-10, 10)
plt.ylim(0, 10)  # same as the x-axis for equality
plt.title('Axis Limits and Tight Scaling')
plt.axis('tight')
plt.show()

plt.plot(x, f_x)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title('Reversed Axes and Equal Scaling')
plt.axis('equal')
plt.show()

```

6. Lables and Legends

```python
x = np.linspace(0, 4, 100)
f_x = np.sqrt(x)
g_x = x**(1/3)

plt.plot(x, f_x, '-g', label='$f(x) = \sqrt{x}$')
plt.plot(x, g_x, ':r', label='$g(x) = x^{1/3}$')

plt.title('Square Root and Cube Root Functions')
plt.xlabel('x')
plt.ylabel('Function Values')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

7. Basic Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.plot(x, y, 'o', color= 'red')
plt.title('plt.plot() Scatter Plot')
plt.show()

plt.scatter(x, y)
plt.title('plt.scatter() Scatter Plot ')
plt.show()
```

7. Customizing Scatter Plots

```python
x = np.linspace(0, 4*np.pi, 100)
f_x = x * np.sin(x)


plt.scatter(x, f_x, s=50, facecolor='green', edgecolor='black', linewidth=2, label='$f(x) = x \sin(x)$')
plt.plot(x, f_x, linestyle='-', linewidth=1)

plt.title('Custom Scatter Plot')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

8. Combined Scatter Plots

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
f_x = x * np.cos(x)

plt.plot(x, f_x, label='$f(x) = x \cos(x)$', color='blue')

scatter_points = np.arange(-2*np.pi, 2*np.pi, np.pi/4)
scatter_values = scatter_points * np.cos(scatter_points)
plt.scatter(scatter_points, scatter_values, color='black', label='Scatter Points')

plt.title('Combined Line and Scatter Plot')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

10. Custom Plot with Alternative Style

```python
plt.style.use(alternative_style)# different style from task 1

x= np.linspace(-10,10,100)
f_x= np.tan(x)
g_x= x*np.exp(-x**2)
i_x= np.arctan(x)

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.plot(x, f_x, '-g', label='$f(x) = \tan(x)$')
plt.plot(x, g_x, ':r', label='$g(x) = x e^{-x^2}$' )
plt.plot(x, i_x,'-.k', label='$i(x) = \arc(x)/x$')

plt.title('Custom Plot with Multiple Functions')
plt.xlabel('x')
plt.ylabel('Function Values')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

11. Style Comparison

```python
plt.style.use('default')

x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1

plt.plot(x, y)
plt.title('Default Style')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1

plt.plot(x, y)
plt.title('seaborn-whitegrid Style' )
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
plt.style.use(alternative_style)

x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1

plt.plot(x, y)
plt.title('Solarize_Light2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Comparision of the Styles

1. Default style: is basic with no grid and the plot is straightforward. It is simple and clear but hard to assess the exact value because of lack of gridlines.
2. seaborn-whitegrid: very clear and grid lines makes it more precise.
3. Solarize_light2: retro-themed with soft grid and beige background. The background color might be disctracting and may contradict with certain color.Mostly used for non-technical presentaion.

Conclusion: Style choice depend on the context but for technical presentations or reports with detailed analysis 'seaborn-whitegrid' is the best option.
