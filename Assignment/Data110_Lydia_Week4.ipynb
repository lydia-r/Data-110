```python
import matplotlib.pyplot as plt # importing important libraries
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
# inputing dataset
offenses = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
offenses_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
offenses_2022vs2023 = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
```

### Task 1: Basic Bar Chart for crime Rates in 2023

```python
plt.bar(offenses, offenses_2023, color= 'blue')
plt.title('Crime Rates per 100,000 People in 2023')
plt.xlabel('Offense Type')
plt.ylabel('Offenses per 100,000 People (2023)')
plt.xticks(rotation=65) # helps the values in x axis be seen clear by rotating 65degree
plt.tight_layout()
plt.show()
```

```python
C = ['lightblue', 'green', 'yellow', 'red', 'orange','blue','grey']
plt.barh(offenses, offenses_2022vs2023, color= C) # ploting horizontal axis
plt.title('Change in Crime Rates (2022 vs. 2023)')
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)')
plt.ylabel('Offense Type')
```

# Why does this bar graph make more sense than Task 1's bar graph?
 * Task 1 doesn't show the progress it lack contenxt about how it is improving or worsening. but task 2 shows the change over time. it shows whether public safetly is improving or deteriorating over time.

 * absolute numbers alone might hide the critical shift happening over time but percentage change shows trend and pattern over time.



# Improve the clarity of your bar graph

```python
bar= plt.barh(offenses, offenses_2022vs2023, color= C)
for bar, change in zip(bar,offenses_2022vs2023 ):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{change}%', va='center', ha='left', fontsize=10)
plt.title('Change in Crime Rates (2022 vs. 2023)')
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)')
plt.ylabel('Offense Type')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.barh(offenses, offenses_2022vs2023, color= 'lightblue') # ploting horizontal axis
plt.title('Change in Crime Rates (2022 vs. 2023)')
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)')
plt.ylabel('Offense Type')
```

# Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023

```python
offenses_2022 = np.array(offenses_2023) / (1 + np.array(offenses_2022vs2023) / 100) # calculating the year 2022
bar_width = 0.40 # assign bar width to avoid overlapping
index = np.arange(len(offenses)) # generating index equal to offense catagories
plt.bar(index, offenses_2022, bar_width, color='violet', label='2022') # ploting first
plt.bar(index + bar_width, offenses_2023, bar_width, color='darkviolet', label='2023') # ploting second bar slightly right
plt.xlabel('Offense Type')
plt.ylabel('Offenses per 100,000 People')
plt.title('Comparison of Crime Rates in 2022 and 2023')
plt.xticks(index + bar_width / 2, offenses, rotation=45, ha='right')
plt.legend()
plt.show()
```
