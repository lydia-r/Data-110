```python
# import libraires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('https://raw.githubusercontent.com/lydia-r/Data-110/refs/heads/main/Project%201/spotifysongs.csv') # loading the dataset
```

##General Correlation Analysis

we can find the relationship with different features using different methods here I have used paiplot and heatmap.

```python
# Relationship between popularity and other song features (e.g., danceability, energy, loudness, etc.)

features = ['popularity', 'danceability', 'energy', 'loudness', 'valence', 'tempo', 'speechiness', 'acousticness','instrumentalness']
data_features = data[features]
correlation_matrix = data_features.corr()
# Display the correlation matrix
print(correlation_matrix)

sns.pairplot(data_features)
plt.suptitle('Pair Plot of Popularity and Audio Features', y=1.02)
plt.tight_layout()
```

we can also pull out and show relationships individualy for further analysis

The first graph shows the strong positive correlation between energy and loudness
The second graph shows the  neutral correlation between popularity and temp
The last one shows weaker negative correaltion between dancablity and popularity

```python
# realtionship between energy and loudness
sns.scatterplot(x='energy', y='loudness', data=data)
plt.title('Energy vs Loudness')
plt.show()
```

```python
# relationship between popularity and instumental
sns.scatterplot(x='tempo', y='popularity', data=data)
plt.title('Popularity vs Tempo')
plt.show()
```

```python
# relationship between popularity and danceability
sns.scatterplot(x='acousticness', y='popularity', data=data)
plt.title('Popularity vs Danceability')
plt.show()
```

```python
# we use correlation Heatmap to show alternative way of seeing how those feature relate

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Popularity and Song Features')
plt.show()
```

### Data visualization

```python
# Top five most popular artists
top_artists= data.groupby('artist').count().sort_values(by='song', ascending= False)['song'][:5]
top_artists.plot.barh()
plt.title('Top five most popular artists')
plt.show()
```

```python
# top five loudest tracks
five_loudest= data[['loudness','song']].sort_values(by='loudness', ascending=True )[:5]
p=sns.barplot(x='loudness', y='song',data= five_loudest)
p.yaxis.tick_right()
plt.title('Top five Loudest songs')
plt.show()
```

```python
# Top artist with the most dancable songs
top_danceable= data[['danceability','song','artist']].sort_values(by='danceability',ascending= False)[:5]
sns.barplot(x='danceability', y='artist', data= top_danceable)
plt.title('Top Artist with the most dancable song')
```

```python
# Top ten artists with instrumental songs
Top_instrumental= data[['instrumentalness','song', 'artist']].sort_values(by= 'instrumentalness', ascending= False)[:10]
sns.barplot(x='instrumentalness',y='artist', data = Top_instrumental)
plt.title('Top ten artists with higest instrumentalness')
```
